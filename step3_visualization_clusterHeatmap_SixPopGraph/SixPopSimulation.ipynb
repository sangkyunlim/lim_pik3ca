```python
x=[1,2,3,4,5,6]
y1=[200, 300, 400, 450, 230, 100]
y2=[200, 300, 400, 450, 330, 300]
y3=[200, 300, 310, 300, 330, 300]
y4=[200, 100, 400, 450, 230, 100]

y_list = [y1, y2, y3, y4]

ScoreList=list([])
for k in range(4):
    M1 = np.mean( [ y_list[k][1], y_list[k][-1]]  )
    M2 = np.mean( [ y_list[k][0], y_list[k][0]] )
    SD = np.std(y_list[k])
    score = (M1-M2)/SD
    ScoreList = ScoreList + [score]
                   

ScoreList = np.around(ScoreList, decimals=2)
```

```python
f, axs = plt.subplots(1,4,figsize=(6,6))

for k in range(4):
    plt.subplot(2, 2, k+1)
    plt.scatter(x,y_list[k], vmin=0, vmax=450, c='brkkkr', s=100)
    plt.xlabel('Population', fontsize=20)
    string = 'Ab#' + str(k+1)
    plt.ylabel(string, fontsize=20)
    plt.ylim([0, 500])
    m=k+1
    plt.title(ScoreList[k], fontsize=20)
    xticks(np.arange(1,7), ('Pop1', 'Pop2', 'Pop3', 'Pop4', 'Pop5', 'Pop6'))

subplots_adjust(left=1, bottom=1, right=2, top=2,
                wspace=0.4, hspace=0.4)

plt.show()
f.savefig('AC.png')
```
